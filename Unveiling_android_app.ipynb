```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/datasets/apps.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Convert to string first
df['Installs'] = df['Installs'].astype(str)

# Remove commas and plus signs
df['Installs'] = df['Installs'].str.replace(',', '', regex=False)
df['Installs'] = df['Installs'].str.replace('+', '', regex=False)

# Convert to numeric
df['Installs'] = df['Installs'].astype(int)
```

```python
# Convert to string
df['Price'] = df['Price'].astype(str)

# Remove '$' symbol
df['Price'] = df['Price'].str.replace('$', '', regex=False)

# Convert to float
df['Price'] = df['Price'].astype(float)
```

```python
df['Rating'].fillna(df['Rating'].mean(), inplace=True)
```

```python
df['Size'].fillna(df['Size'].median(), inplace=True)
```

```python
df['Current Ver'].fillna("Unknown", inplace=True)
df['Android Ver'].fillna("Varies with device", inplace=True)
```

```python
df.info()
df.isnull().sum()  # Shows remaining missing values, if any
```

```python
df.to_csv("cleaned_apps.csv", index=False)
```

```python
df['Category'].unique()
```

```python
category_counts = df['Category'].value_counts()
print(category_counts)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
top10 = category_counts.head(10)

sns.barplot(x=top10.index, y=top10.values, palette="viridis")
plt.title("Top 10 App Categories by Count")
plt.xlabel("Category")
plt.ylabel("Number of Apps")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 10))
df['Category'].value_counts().plot.pie(autopct='%1.1f%%', startangle=140, colormap='Set3')
plt.title("App Distribution by Category")
plt.ylabel("")  # hide y-label
plt.show()
```

```python
free_paid_counts = df.groupby(['Category', 'Type']).size().unstack(fill_value=0)
free_paid_counts.head()
```

```python
free_paid_counts_sorted = free_paid_counts.sort_values(by='Free', ascending=False).head(10)

free_paid_counts_sorted.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='Paired')
plt.title("Free vs Paid Apps by Category (Top 10)")
plt.xlabel("Category")
plt.ylabel("Number of Apps")
plt.xticks(rotation=45)
plt.legend(title="App Type")
plt.tight_layout()
plt.show()
```

```python
avg_rating = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)
avg_rating.head(10)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_rating.head(10).values, y=avg_rating.head(10).index, palette="coolwarm")
plt.title("Top 10 Categories by Average Rating")
plt.xlabel("Average Rating")
plt.ylabel("Category")
plt.tight_layout()
plt.show()
```

```python
reviews_df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/datasets/user_reviews.csv")
reviews_df.head()
```

```python
reviews_df.dropna(subset=['Translated_Review', 'Sentiment'], inplace=True)
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=reviews_df, x='Sentiment', palette='Set2')
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Number of Reviews")
plt.show()
```

```python
from wordcloud import WordCloud

for sentiment in ['Positive', 'Negative', 'Neutral']:
    text = " ".join(review for review in reviews_df[reviews_df['Sentiment'] == sentiment]['Translated_Review'])
    wc = WordCloud(width=800, height=400, background_color='white').generate(text)

    plt.figure(figsize=(10, 5))
    plt.imshow(wc, interpolation='bilinear')
    plt.axis('off')
    plt.title(f"Word Cloud for {sentiment} Reviews")
    plt.show()
```

```python
import plotly.express as px
```

```python
fig = px.scatter(
    df,
    x="Rating",
    y="Installs",
    size="Reviews",
    color="Category",
    hover_name="App",
    title="App Rating vs Installs (Bubble Size = Reviews)",
    size_max=60
)
fig.show()
```

```python
category_counts = df['Category'].value_counts().reset_index()
category_counts.columns = ['Category', 'App Count']

fig = px.bar(
    category_counts,
    x='Category',
    y='App Count',
    title='Number of Apps per Category',
    color='App Count',
    color_continuous_scale='Viridis'
)
fig.show()
```

```python
fig = px.box(
    df,
    x='Type',
    y='Rating',
    color='Type',
    title="Rating Distribution for Free vs Paid Apps"
)
fig.show()
```

```python
sentiment_grouped = reviews_df.groupby(['App', 'Sentiment']).size().reset_index(name='Review Count')

fig = px.bar(
    sentiment_grouped,
    x='App',
    y='Review Count',
    color='Sentiment',
    title='Sentiment Distribution Across Apps'
)
fig.show()
```
